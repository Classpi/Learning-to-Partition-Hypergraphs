```python
import torch
import dhg
import kahypar
from dhg.data import Cooking200,HouseCommittees
from hgp.utils import from_hypergraph_to_kahypar,from_pickle_to_kahypar
import pandas as pd

```

```python
df = pd.DataFrame(columns=["k=2","k=3","k=4","k=5","k=6"])
```

```python
dataset = [Cooking200(),HouseCommittees()]
for data in dataset:
    df.loc[data.name] = [None] * (len(df.columns))
    for k in range(2, 7):
        num_nodes, num_nets, e_index, e_list, edge_weights, node_weights = from_hypergraph_to_kahypar(data)
        context = kahypar.Context()
        context.loadINIconfiguration("../config/km1_kKaHyPar_sea20.ini")
        context.setEpsilon(0.01)
        hypergraph = kahypar.Hypergraph(num_nodes, num_nets, e_index, e_list, k, edge_weights, node_weights)
        context.setK(k)
        partition = kahypar.partition(hypergraph, context)
        df.at["{}".format(data.name), "k={}".format(k)] = kahypar.cut(hypergraph)
    
```
